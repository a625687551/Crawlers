```python
import pandas as pd
import numpy as np
# import traceback
# traceback.print_exc(file='123.txt')
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
%pylab
# import krisk.plot as kk
# from krisk import init_notebook;init_notebook
```

```python
df=pd.read_csv('wea.txt',names=['city','AQI'])
df.head()
```

```python
df.set_index('city',inplace=True)
```

```python
m=Basemap(width=7000000,height=5000000,projection='merc',
         resolution='c',lat_1=20.,lat_2=50,lat_0=35,lon_0=110.)
```

```python
m.drawcountries()
m.drawcoastlines()
parallels=np.arange(0,81,10)
meridians=np.arange(70,140,20)
m.drawparallels(parallels,labels=[False,True,True,False])
m.drawmeridians(meridians,labels=[True,False,False,True])
plt.show()
```
